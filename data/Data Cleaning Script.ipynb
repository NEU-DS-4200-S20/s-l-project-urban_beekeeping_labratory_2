```python
df = pd.read_csv("MassData.csv")
df
```

```python
#filtering to keep data records where not all 3 counts are null
df_clean_1 = df[df.BeeCount.notnull() | df.BroodCount.notnull() | df.HoneyCount.notnull()]
df_clean_1
```

```python
#filtering to get rid of any data points with a "?" count present
df_clean_2 = df_clean_1[(df_clean_1.BeeCount != "?")& (df_clean_1.BroodCount != "?") & (df_clean_1.HoneyCount != "?")]
df_clean_2
```

```python
#filtering any data point where all 3 counts are 0
df_clean_3 = df_clean_2[(df_clean_2.BeeCount != "0") & (df_clean_2.BroodCount != "0") & (df_clean_2.HoneyCount != "0")]
df_clean_3
```
